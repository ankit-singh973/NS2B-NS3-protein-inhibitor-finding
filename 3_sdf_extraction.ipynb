## **Aim of this notebook is to download SDF files of all the 212 molecules from drug bank for virtual screening**

```python
import pandas as pd
```

```python
act_df = pd.read_csv("active_molecules_1.csv")
```

```python
act_df
```

```python
ids = act_df['DrugBank ID'].to_list()
```

```python
ids
```

```python
from rdkit import Chem
from rdkit.Chem import AllChem
import requests
import os

# List of DrugBank IDs
drugbank_ids = ids

# Function to download SDF file for a DrugBank ID
def download_sdf(drugbank_id):
    url = f'https://go.drugbank.com/structures/small_molecule_drugs/{drugbank_id}.sdf'
    # url = f'https://go.drugbank.com/structures/{drugbank_id}/download_sdf'
    response = requests.get(url)
    if response.status_code == 200:
        if not os.path.exists('SDFs'):
            os.makedirs('SDFs')
        with open(f'SDFs/{drugbank_id}.sdf', 'wb') as file:
            file.write(response.content)
            print(f'{drugbank_id} SDF file downloaded successfully')
    else:
        print(f'Failed to download SDF file for {drugbank_id}')

# Download SDF files for each DrugBank ID
for drugbank_id in drugbank_ids:
    download_sdf(drugbank_id)
```
